```python
df_t=df_train.drop(['id','purchaseTime','hour','N10','N8'] , axis=1)
```

```python
df_t.head(2)
```

```python
X = df_t.loc[:, df_t.columns != "label"]
y = df_t.loc[:, df_t.columns == "label"]
```

```python
!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 42)
```

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(sampling_strategy = 'minority', random_state = 42)
X_train, y_train= sm.fit_resample(X_train, y_train)
```

```python
len(y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42, n_estimators=60)

rfc.fit(X_train, y_train)
pred_rfc = rfc.predict_proba(X_test)
y_pred_rfc = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_rfc))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_rfc))
```

```python
from sklearn.model_selection import cross_val_score
from numpy import mean
scores = cross_val_score(rfc, X, y, scoring='roc_auc',n_jobs=-1)
print('Mean ROC AUC: %.3f' % mean(scores))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)

print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
